```python
def makeMagnitudeList():
    quakefile = open('earthquakes.txt','r')
    maglist = []
    for aline in quakefile :
        vlist = aline.split()
        mag = vlist[1]
        maglist.append(mag)
    quakefile.close()
    return maglist
```

```python
def countHead(url):
    page = urllib.request.urlopen(url)
    line = page.readline().decode('utf-8')
    while '<head>' not in line :
        line = page.readline().decode('utf-8')
    n_line = 1
    while '</head>' not in line :
        line = page.readline().decode('utf-8')
        n_line = n_line + 1
    print('머릿말의 줄 수: %d' %n_line)
```

```python
url = 'http://www.naver.com/index.html'
```
